```python
import torch
import numpy as np
import train_models

directory = "C:/Users/lrozier/Documents/UQAC/respiratory-disease-detection/data/processed/"

# Configuration RTX 4070 : Tensor Cores (TF32)
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Propulsé par : {torch.cuda.get_device_name(0)}\n")
```

```python
print ("\nEtude du modèle CNN ...\n\n")
models_ccn = []
history_ccn = []
accuracy_ccn = []
macro_f1_ccn = []
mean_auc_ccn = []
for _ in range(5):
    model, history, accuracy, macro_f1, mean_auc = train_models.run_model("CNN", directory, device)
    models_ccn.append(model)
    history_ccn.append(history)
    accuracy_ccn.append(accuracy)
    macro_f1_ccn.append(macro_f1)
    mean_auc_ccn.append(mean_auc)
```

```python
print ("\nEtude du modèle ConvNeXt ...\n\n")
models_convnext = []
history_convnext = []
accuracy_convnext = []
macro_f1_convnext = []
mean_auc_convnext = []
for _ in range(5):
    model, history, accuracy, macro_f1, mean_auc = train_models.run_model("ConvNeXt", directory, device)
    models_convnext.append(model)
    history_convnext.append(history)
    accuracy_convnext.append(accuracy)
    macro_f1_convnext.append(macro_f1)
    mean_auc_convnext.append(mean_auc)
```

```python
print ("\nEtude du modèle ConvNeXt_with_features ...\n\n")
models_convnext_with_features = []
history_convnext_with_features = []
accuracy_convnext_with_features = []
macro_f1_convnext_with_features = []
mean_auc_convnext_with_features = []
for _ in range(5):
    model, history, accuracy, macro_f1, mean_auc = train_models.run_model("ConvNeXt_with_features", directory, device)
    models_convnext_with_features.append(model)
    history_convnext_with_features.append(history)
    accuracy_convnext_with_features.append(accuracy)
    macro_f1_convnext_with_features.append(macro_f1)
    mean_auc_convnext_with_features.append(mean_auc)
```

```python
print("="*70)
print("\t\tRÉSULTAT DE L'ÉVALUATION DES MODÈLES")
print("="*70)

def print_3x3(matrice):
    format_row = "| {:^8} | {:^8} | {:^8} | {:^15} |"
    sep = "+" + "----------+" * 3 + "-----------------+"
    
    print(sep)
    for row in matrice:
        print(format_row.format(*row))
        print(sep)



grid_mean = [["", "CNN", "ConvNeXt", "ConvNeXt_feats"], 
        ["Accuracy",np.mean(accuracy_ccn).round(2), np.mean(accuracy_convnext).round(2) , np.mean(accuracy_convnext_with_features).round(2)], 
        ["Macro F1",np.mean(macro_f1_ccn).round(2) , np.mean(macro_f1_convnext).round(2), np.mean(macro_f1_convnext_with_features).round(2)], 
        ["AUC",np.mean(mean_auc_ccn).round(2), np.mean(mean_auc_convnext).round(2), np.mean(mean_auc_convnext_with_features).round(2)]]


print_3x3(grid_mean)

print("\n\n")

grid_max = [["", "CNN", "ConvNeXt", "ConvNeXt_feats"],
        ["Accuracy",np.max(accuracy_ccn).round(2), np.max(accuracy_convnext).round(2) , np.max(accuracy_convnext_with_features).round(2)], 
        ["Macro F1",np.max(macro_f1_ccn).round(2) , np.max(macro_f1_convnext).round(2), np.max(macro_f1_convnext_with_features).round(2)], 
        ["AUC",np.max(mean_auc_ccn).round(2), np.max(mean_auc_convnext).round(2), np.max(mean_auc_convnext_with_features).round(2)]]

print_3x3(grid_max)
```
